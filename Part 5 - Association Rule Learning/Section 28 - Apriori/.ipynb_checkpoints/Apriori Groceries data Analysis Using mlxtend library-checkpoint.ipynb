## Importing common libraries

```python
import numpy as np 
import pandas as pd 
```

```python
pd.set_option('display.max_columns', 999)
pd.set_option('display.max_rows', 999)
```

```python
data = pd.read_excel('Australia Online Retail.xlsx') 
data.head() 
```

## Cleaning the Data

```python
# Stripping extra spaces in the description 
data['Description'] = data['Description'].str.strip() 
```

```python
data.isnull().sum()
```

## Splitting the data according to transaction

```python
data = (data.groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo'))
```

```python
data.head()
```

## Defining the hot encoding function to make the data suitable  
## for the concerned libraries 

```python
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
```

```python
data = data.applymap(hot_encode)
```

```python
data.head()
```

## Buliding the models and analyzing the results

```python
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
frq_items = apriori(data, min_support = 0.05, use_colnames = True) 
```

## Collecting the inferred rules in a dataframe 

```python
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]).reset_index()
rules.head()
```
