```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns 

import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
warnings.filterwarnings(action='ignore', category=FutureWarning)
```

```python
input_DF = pd.read_csv(filepath_or_buffer='winequality-red.csv', sep=';')
X = input_DF.loc[:, input_DF.columns != 'quality']
y = input_DF[['quality']]
```

# Checking problem is a linear or nonlinear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
test_for_normality_model = LinearRegression()
test_for_normality_model.fit(X,y)
print('R2 for the test model', r2_score(y_true=y, y_pred=test_for_normality_model.predict(X)))
print('RMSE for the test model', np.sqrt(mean_squared_error(y_true=y, y_pred=test_for_normality_model.predict(X))))
```

### RMSE value is good enough to consider this as a linear problem

```python
from scipy.stats import shapiro
```

```python
pvalue = shapiro(test_for_normality_model.predict(X))[1]
print('pvalue > .5 : ', (pvalue > .5) )
```

### Statistically no eidence to reject null hypothesis, its a normal distribution
### Both tests indicate its linear problem

# Goodness of test with backward elimination with p value & Adj-R2

```python
from statsmodels.regression.linear_model import OLS
from statsmodels.tools.tools import add_constant
```

```python
def backward_elimination(X, y, significance_level):
    cols = list(X.columns)
    pmax = 1
    previous_adj_r2 = 1
    iteration_count = 0
    
    while (len(cols)>0):
        p= []
        X_1 = X[cols]
        X_1 = add_constant(X_1)
        model = OLS(y,X_1).fit()
        p = pd.Series(model.pvalues.values[1:],index = cols) 
        pmax = max(p)
        feature_with_p_max = p.idxmax()
        if(iteration_count > 0):
            if(previous_adj_r2 > model.rsquared_adj):
                break;
            else:
                previous_adj_r2 = model.rsquared_adj
                if(pmax > significance_level):
                    cols.remove(feature_with_p_max)
                else:
                    break;
        else:
            previous_adj_r2 = model.rsquared_adj
            if(pmax > significance_level):
                cols.remove(feature_with_p_max)
            else:
                break;
    return cols, X[cols]
```

```python
important_cols, new_X = backward_elimination(X, y, significance_level=.4)
X = new_X
important_cols
```

# Checking for Independent features / No multi collinearity

```python
corr = X.corr()
plt.pyplot.subplots(figsize=(10,5))
mask = np.triu(np.ones_like(corr, dtype=np.bool))
ax = sns.heatmap(data=corr, annot=True, cmap="YlGnBu", mask=mask)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

### satidfied that there is no multicolinearity issue as all correlation values are < |.7| value

# Create Test Train splits

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
y_train = sc_y.fit_transform(y_train)
y_test = sc_y.transform(y_test)
```

# Create simple SVR model

```python
from sklearn.svm import SVR
```

```python
simple_regressor = SVR(kernel='rbf')
simple_regressor.fit(X_train, y_train)
```

# Simple model performance evaluation

```python
def adjusted_r2(y_actual, y_pred, sample_size, feature_count):
    from sklearn.metrics import r2_score
    R2 = r2_score(y_actual, y_pred)
    n = sample_size
    p = feature_count
    Adj_r2 = 1-(1-R2)*(n-1)/(n-p-1)
    return Adj_r2
```

### RMSE & Adj-R2 for the simple model on Test data

```python
print('RMSE of Simple model', np.sqrt(mean_squared_error(y_true=y_test, y_pred=simple_regressor.predict(X_test))))
print('Adj-R2 of Simple model', adjusted_r2(y_test, simple_regressor.predict(X_test), len(X_test), len(X.columns)))
```

# Hyper Tunning the model for better performance

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [ {'C' : [.8, 1, 1.2, 1.3, 1.5], 
                 'epsilon' : [.001, .1, .3, .5], 
                 'degree' : [1,3,5,8],
                'gamma' : ['scale','auto']}]  
```

```python
regressor=SVR(kernel='rbf')
search = GridSearchCV(estimator=regressor, param_grid=parameters, cv=5, n_jobs=-1)
search.fit(X=X_train, y=y_train)
best_rbf_regressor = search.best_estimator_
```

### RMSE & Adj-R2 for the Hypter Tunned model on Test data

```python
print('RMSE of Hyper Tunned model', np.sqrt(mean_squared_error(y_true=y_test, y_pred=best_rbf_regressor.predict(X_test))))
print('Adj-R2 of Hyper Tunned model', adjusted_r2(y_test, best_rbf_regressor.predict(X_test), len(X_test), len(X.columns)))
```

# Hyper tuned parameters with K-Fold Training

```python
from sklearn.model_selection import KFold
```

```python
rkf = KFold(n_splits=5, random_state=0)
regressor_kf = SVR(C=0.8, cache_size=200, coef0=0.0, degree=1, epsilon=0.3, gamma='scale',
    kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)

for train_index, test_index in rkf.split(X_train):
    X_train_n, X_test_n = X_train[train_index], X_train[test_index]
    y_train_n, y_test_n = y_train[train_index], y_train[test_index]
    regressor_kf.fit(X_train_n, y_train_n)
```

### RMSE & Adj-R2 for the K-Fold Trained Hypter Tunned model on Test data

```python
print('RMSE of Simple model', np.sqrt(mean_squared_error(y_true=y_test, y_pred=regressor_kf.predict(X_test))))
print('Adj-R2 of Simple model', adjusted_r2(y_test, regressor_kf.predict(X_test), len(X_test), len(X.columns)))
```

# Simple model with K-Fold Training

```python
rkf = KFold(n_splits=5, random_state=0)
regressor_kfs = SVR( kernel='rbf')

for train_index, test_index in rkf.split(X_train):
    X_train_n, X_test_n = X_train[train_index], X_train[test_index]
    y_train_n, y_test_n = y_train[train_index], y_train[test_index]
    regressor_kfs.fit(X_train_n, y_train_n)
```

### RMSE & Adj-R2 for the K-Fold Trained Simple model on Test data

```python
print('RMSE of Simple model', np.sqrt(mean_squared_error(y_true=y_test, y_pred=regressor_kfs.predict(X_test))))
print('Adj-R2 of Simple model', adjusted_r2(y_test, regressor_kfs.predict(X_test), len(X_test), len(X.columns)))
```
