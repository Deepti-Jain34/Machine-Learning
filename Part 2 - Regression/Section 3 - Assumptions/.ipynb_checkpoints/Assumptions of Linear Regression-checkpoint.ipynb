## Import libraries

```python
%matplotlib inline
%config InlineBackend.figure_format ='retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns',80)
sns.set_style('darkgrid')
sns.mpl.rcParams['figure.figsize'] = (12.0, 6.0)
```

## load data and split to depend & indipend varriables X & y respectiely

```python
data = pd.read_csv(filepath_or_buffer=r'.\housesales\kc_house_data.csv')
y = data[['price']]
X = data[['bedrooms','bathrooms','sqft_living','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode']]
X.head()
```

## Run a Linear Regression

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from statsmodels.tools.tools import add_constant
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=0)
model = sm.OLS(exog=add_constant(x_train), endog=y_train.to_numpy()).fit()
```

## 1. Check for linearity

### 1) Scatter plots

```python
def plot_linearity_scatter(data, indipend_cols, depend_col, row, col):
    column_index=0
    fig, ax = plt.subplots(row,col)
    for i in range(row):
        for j in range(col):
            sns.scatterplot(x=indipend_cols[column_index], y=depend_col, ax=ax[i,j], data=data)
            ax[i,j].set_title(indipend_cols[column_index]+' '+depend_col)
            ax[i,j].set(xlabel=indipend_cols[column_index], ylabel=depend_col)
            column_index += 1
    plt.tight_layout()
    plt.show()
```

```python
plot_linearity_scatter(data=data, indipend_cols=list(X.columns), depend_col='price', row=2, col=4)
```

## 2. Normality of reciduals

#### 1) QQ plot

```python
sm.ProbPlot(model.resid).qqplot(line='s');
plt.title('Q-Q plot');
```

### 2) tests

```python
from scipy import stats
jb = stats.jarque_bera(model.resid)
sw = stats.shapiro(model.resid)
ad = stats.anderson(model.resid, dist='norm')
ks = stats.kstest(model.resid, 'norm')
```

```python
print(f'1) Jarque-Bera test ---- statistic: {jb[0]:.4f}, p-value: {jb[1]}')
print(f'2) Shapiro-Wilk test ---- statistic: {sw[0]:.4f}, p-value: {sw[1]:.4f}')
print(f'3) Kolmogorov-Smirnov test ---- statistic: {ks.statistic:.4f}, p-value: {ks.pvalue:.4f}')
print(f'4) Anderson-Darling test ---- statistic: {ad.statistic:.4f}, 5% critical value: {ad.critical_values[2]:.4f}')
print('5) If the returned AD statistic is larger than the critical value, then for the 5% significance level','\n the null hypothesis that the data come from the Normal distribution should be rejected. ')
```

#### As sample numbers are > 5000 test results cant be trusted
#### But According to both plot & test Recedues are not normally distributed

## 3. Multicolinearity

```python
model.summary()
```

### 1) Calculate condition number

#### 1] The first step is to normalize the independent variables to have unit length:

```python
norm_x = x_train.values
for i, name in enumerate(x_train):
    if name == "const":
        continue
    norm_x[:,i] = x_train[name]/np.linalg.norm(x_train[name])
norm_xtx = np.dot(norm_x.T,norm_x)
```

#### 2] Then, take the square root of the ratio of the biggest to the smallest eigen values.

```python
eigs = np.linalg.eigvals(norm_xtx)
condition_number = np.sqrt(eigs.max() / eigs.min())
print(condition_number)
```

#### As cont. no greater than 30 multicolineaarity is present

### 2) VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X_constant = add_constant(X)
vif = [variance_inflation_factor(X_constant.values, i) for i in range(X_constant.shape[1])]
pd.DataFrame({'vif': vif[1:]}, index=X.columns).T
```

#### There is multi colinearity present

## 4) Homoscedasticity

### 1) plot

```python
predicted_vals = model.predict().reshape(-1,)
resids = model.resid.to_numpy().reshape(-1,)
y_obsered = y_train.to_numpy().reshape(-1,)
```

```python
fig, ax = plt.subplots(1,2)
sns.regplot(x=predicted_vals, y=y_obsered, lowess=True, ax=ax[0], line_kws={'color': 'red'})
ax[0].set_title('Observed vs. Predicted Values', fontsize=16)
ax[0].set(xlabel='Predicted', ylabel='Observed')
sns.regplot(x=predicted_vals, y=resids, lowess=True, ax=ax[1], line_kws={'color': 'red'})
ax[1].set_title('Residuals vs. Predicted Values', fontsize=16)
ax[1].set(xlabel='Predicted', ylabel='Residuals')
```

#### funnel shape pattern in Residual vs. Fitted Values Plot suggests data is suffering from heteroskedasticity

### 2) breusch pagan & goldfeld quand tests

```python
import statsmodels.stats.api as sms
```

```python
breuschpagan_result = sms.het_breuschpagan(model.resid, add_constant(x_train))
bp_test = pd.DataFrame(data = breuschpagan_result, 
                           columns = ['value'],
                           index = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value'])
```

```python
goldfeldquand_result = sms.het_goldfeldquandt(model.resid, model.model.exog)
gq_test = pd.DataFrame(data = goldfeldquand_result[:-1],
                           columns=['value'],
                           index=['F statistic', 'p-value'])
```

```python
print('\n Breusch-Pagan test ----')
bp_test.head()
```

```python
print('\n Goldfeld-Quandt test ----')
gq_test.head()
```

### as the result of Breusch-Pagan tests found p < 0.05, thus reject the null hypothesis and infer that heteroskedasticity is present.
